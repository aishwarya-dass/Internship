# EXPLOLATORY DATA ANALYSIS (EDA)
### Dataset : SampleSuperstore
## OBJECTIVE :
#### 1.To find out the weak areas we need to work at in order to make profits
#### 2.Business Problem we can derive from this EDA


```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```
x=pd.read_csv('SampleSuperstore.csv')
x
```

```
per=pd.DataFrame(x)
print(per)
```

###Data Cleaning

```
per.head()
```

```
per.tail()
```

```
per.nunique()
```

```
duplicate = per[per.duplicated()]
print(duplicate)
```

```
per.drop_duplicates(inplace=True)
```

### Postal code is not used for analysis ,hence we can drop postal code.

```
per.drop('Postal Code',axis=1,inplace=True)
```

```
per
```

```
per.describe()
```

```
per.corr()
```

###Data Analysis

###Analyzing profits in a hierarchical way
### In terms of location
#####1.Profit vs Region
#####2.Profit vs State
#####3.Profit vs City

### In terms of Product
#####1.Profit vs Category
#####2.Profit vs Sub Category

```
sns.barplot(x='Profit',y='Region',data=per)
plt.show()
```

```
plt.figure(figsize=(10,15))
ax=sns.barplot(x='Profit',y='State',data=per)
```

```
plt.figure(figsize=(30,100))
ax=sns.barplot(x='Profit',y='City',data=per)
```

```
sns.barplot(x='Profit',y='Category',data=per)
plt.show()
```

```
sns.barplot(x='Profit',y='Sub-Category',data=per)
plt.show()
```

###Establishing relationship between Discount and Profit

```
sns.barplot(x='Discount',y='Profit',data=per)
plt.show()
```

###Establishing relationship between Mode of shipment and profit

```
sns.barplot(x='Ship Mode',y='Profit',data=per)
plt.show()
```
